<h2 class="text-3xl font-bold text-purple-800">My CSSE Journey and Final Project: <em>The End RPG Game</em></h2>
<p>Throughout CSSE1 and CSSE2, especially from Trimester 2 to 3, I built a strong foundation in coding by learning core programming skills like JavaScript, control structures, object-oriented programming, and software engineering practices.</p>
<p>I worked closely with my team to plan, design, and develop a 2D RPG game called <strong>The End</strong>, which serves as our final project.</p>
<p><strong>The End</strong> is more than just a game—it’s a showcase of everything we’ve learned. We implemented features like finite state machines for game flow, player inventory systems, interactive NPCs, and dynamic environments using object classes. I contributed by debugging complex issues, writing modular code, and collaborating on gameplay logic and design.</p>

<h2 class="mt-6 text-2xl font-semibold">Software Engineering Practices</h2>
<ul class="list-disc ml-6">
<li><strong>Planning Changes</strong>: Organizing updates or features before writing code.</li>

<ul class="list-disc ml-6">
<li><strong>Source Control</strong>: Managing code changes with Git.</li>
<li><strong>Forking</strong>: Making personal repo copies for changes.</li>
<li><strong>Branching</strong>: Developing features or fixes safely.</li>
<li><strong>Building</strong>: Compiling/preparing the program to run.</li>
<li><strong>Testing and Verification</strong>: Making sure code works through testing.</li>
<li><strong>Pull Requests</strong>: Requesting to merge reviewed changes.</li>
<li><strong>Merging</strong>: Combining branches.</li>

</ul>
